```python
import os
import shutil
!pip install pandas
import pandas as pd

# Define paths
csv_file_path = 'C:/Hacks/images.csv'  # Path to your CSV file
images_dir = 'C:/Hacks/images_compressed'  # Path to your directory containing images
output_dir = 'C:/Hacks/shirt'  # Path to the directory where you want to store sorted images

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Load CSV file
# Assuming CSV columns are: 'image_id' and 'label_name'
df = pd.read_csv(csv_file_path)

# Get unique label names
label_names = df['label'].unique()

# Process each label one at a time
for label_name in label_names:
    print(f"Processing label: {label_name}...")

    # Create the label-specific directory if it doesn't exist
    label_dir = os.path.join(output_dir, str(label_name))
    if not os.path.exists(label_dir):
        os.makedirs(label_dir)

    # Filter the DataFrame for images belonging to the current label
    label_df = df[df['label'] == label_name]

    # Move images of the current label to the label-specific folder
    for index, row in label_df.iterrows():
        image_id = row['image']  # Get the image ID

        # Source image path
        src_image_path = os.path.join(images_dir, f"{image_id}.jpg")  # Adjust file extension as needed

        # Destination image path
        dest_image_path = os.path.join(label_dir, f"{image_id}.jpg")

        # Move or copy the image to the label-specific folder
        if os.path.exists(src_image_path):
            shutil.move(src_image_path, dest_image_path)  # You can use shutil.copy if you want to copy instead of move
        else:
            print(f"Warning: Image {image_id}.jpg not found!")

    print(f"Label {label_name} images have been moved to {label_dir}.\n")

print("All labels have been processed.")
```

```python
import os
import numpy as np
from PIL import Image
import albumentations as A

# Paths to the dataset - List of 11 input folders
input_dirs = [
    
    'C:/Hacks/dataset/train/Pants',
    'C:/Hacks/dataset/train/Shirt',
    'C:/Hacks/dataset/train/Shoes',
    'C:/Hacks/dataset/train/Shorts',
    'C:/Hacks/dataset/train/tshirt',
    'C:/Hacks/dataset/train/Undershirt'
]

# Corresponding output directories for each input folder
output_dirs = [
    
    'C:/Hacks/dataset/test/Pants',
    'C:/Hacks/dataset/test/Shirt',
    'C:/Hacks/dataset/test/Shoes',
    'C:/Hacks/dataset/test/Shorts',
    'C:/Hacks/dataset/test/tshirt',
    'C:/Hacks/dataset/test/Undershirt'
]

# Create output directories if they don't exist
for output_dir in output_dirs:
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

# Define augmentation techniques using albumentations
augmenter = A.Compose([
    A.HorizontalFlip(p=0.5),  # Flip 50% of images horizontally
    A.VerticalFlip(p=0.2),  # Flip 20% of images vertically
    A.Rotate(limit=25, p=0.5),  # Rotate images between -25 and 25 degrees
    A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),  # Adjust brightness/contrast
    A.GaussianBlur(blur_limit=3, p=0.3),  # Apply Gaussian blur
    A.RandomCrop(width=64, height=64, p=0.5),  # Randomly crop the image
])

# Function to augment and save images
def augment_images(input_dir, output_dir, augmenter, num_augmented=5):
    for filename in os.listdir(input_dir):
        if filename.endswith('.jpg') or filename.endswith('.png'):  # Add other file types if needed
            # Load image
            image_path = os.path.join(input_dir, filename)
            image = Image.open(image_path)
            image_np = np.array(image)  # Convert PIL image to numpy array

            # Apply augmentations
            for i in range(num_augmented):
                augmented = augmenter(image=image_np)  # Augment the image
                augmented_image_np = augmented['image']

                # Convert numpy array back to a PIL image
                augmented_image_pil = Image.fromarray(augmented_image_np)

                # Save augmented image with a new name
                output_filename = f"{os.path.splitext(filename)[0]}_augmented_{i+1}.jpg"
                output_image_path = os.path.join(output_dir, output_filename)
                augmented_image_pil.save(output_image_path)

            print(f"Augmented {filename} from {input_dir} and saved to {output_dir}")

# Augment dataset for all 11 folders
for input_dir, output_dir in zip(input_dirs, output_dirs):
    augment_images(input_dir, output_dir, augmenter, num_augmented=5)  # 5 augmented versions per image
```
